# Análise dos cruzeiros pelo mundo

Apenas para treinar um pouco de Pandas

```python
import pandas as pd
```

Importar os arquivos do data.world relativos aos cruzeiros. Excelente perfil pegou o material: https://data.world/brandon-telle/cruise-ship-locations

```python
cities = pd.read_csv('https://query.data.world/s/54jlpsxer4xmbxpwfte5ddha2')
```

```python
location = pd.read_csv('https://query.data.world/s/2e5uh0bxh7b2jcw6gs5hf4xqx')
```

```python
ships = pd.read_csv('https://query.data.world/s/2j7s4x3hx6s41mi8oc7mbvlc')
```

Os cruzeiros só param em três cidades brasileiras.

```python
cities[cities.country=="Brazil"].head()
```

Há 3566 paradas nessas cidades. 

```python
location[location.port_city_country=="BR"]
```

Vamos ver o número de paradas no Brasil por ano. O pico foi em 2011, ano bem otimista ainda. Mas 2015 e 2016 não foram mal. 

```python
year = location.date.str[:4]
```

```python
year[location.port_city_country=="BR"].value_counts()
```

Quantas paradas em cada cidade brasileira

```python
location[location.port_city_country=="BR"].port_city_name.value_counts()
```

Ranking das cidades no mundo; Rio é a nonagésima cidade com mais paradas. Santos é a nonagésima-quinta

```python
location.port_city_name.value_counts().head(100).reset_index()
```

Paradas no Brasil por empresa

```python
location.major_cruise_line[location.port_city_country=="BR"].value_counts()
```

Paradas no Brasil por navio; o mais frequente era bem famoso: https://pt.wikipedia.org/wiki/MS_Splendour_of_the_Seas

```python
location.ship_name[location.port_city_country=="BR"].value_counts()
```

```python
ships.head()
```

O que faz sentido: http://www.nomoretax.eu/living-in-international-waters-endless-adventure-without-taxes/

```python
ships.flag.value_counts()
```

```python
ships.home_port.value_counts()
```

Unir as tabelas ships e location. 

```python
merged = pd.merge(ships,location,on="callsign")
```

```python
merged.head()
```

```python
merged.flag[merged.port_city_country=="BR"].value_counts()
```

Média dos tamanhos dos navios, no Brasil e por cidade.

```python
merged[merged.port_city_country=="BR"].groupby("port_city_name").length.mean().reset_index()
```

Média da distância viajada até o porto, no Brasil e por cidade. 

```python
merged[merged.port_city_country=="BR"].groupby("port_city_name").meters_traveled.mean().reset_index()
```

Média da capacidade do navio, no Brasil e por cidade.

```python
merged[merged.port_city_country=="BR"].groupby("port_city_name").capacity.mean().reset_index()
```

Média de minutos parado no porto, no Brasil e por cidade.

```python
merged[merged.port_city_country=="BR"].groupby("port_city_name").stopped_minutes.mean().reset_index()
```

```python
merged[merged.port_city_country=="BR"].groupby(["major_cruise_line_x","port_city_name"]).stopped_minutes.mean().reset_index().sort_values("stopped_minutes", ascending=False)
```
